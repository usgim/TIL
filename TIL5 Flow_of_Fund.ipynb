### Step 5. Date column을 index로 할당하세요

### Step 7. index를 DatetimeIndex type으로 변경하세요

```python
df_date.index = pd.to_datetime(df_date.index)
```

```python
df_date.index # 시계열 데이터로 전환
```

### Step 8.  frequency 를 월 단위로 변경하고(월말 기준), 각 column의 월단위로 합산값을 구하세요

```python
#resample 사용, Column이나 Index에 접근하지 않아도 전체 df_date에서 .resample 적용
df_m = df_date.resample("M").sum()

# df_m
```

```python
df_m[df_m != 0]
```

### Step 9. Step8에서의 결과중 NaN으로 채워진 값이 존재하는 row를 삭제 하세요

```python
df_m = df_m[df_m != 0].dropna()

# [df_m != 0]는 df_m에 있는 0의 값을 결측치(false)로 만들어
# df_m[df_m != 0]로 인덱싱할 때 NaN으로 변환하게하여
# 행과 열에 NaN이 존재할 경우 제거하는 dropna를 사용한다.
# 최종적으로 df_m = df_m[df_m != 0].dropna()하여 df_m의 값을 수정.
```
