### Step 5. 문자열의 앞글자를 대문자로 변경해주는 labda 함수 또는 함수를 만드세요

```python
def upper1(str):
  return str.capitalize() # capitalize()는 앞 글자를 대문자로 변환해주는 메서드.
```

### Step 6. Mjob 와 Fjob column에 step5 에서 만든 function을 적용하고 원래 stud_alcoh DataFrame에 적용하세요

```python
# applymap : 각 columns마다 돌아다니면서 다 적용
df['Mjob'] = df['Mjob'].apply(upper1) # apply는 각 열에 함수를 적용
df['Fjob'] = df['Fjob'].apply(upper1) # df의 "Fjob"열에 upper1 함수를 적용한 새로운 Fjob의 데이터를 입력하겠다.
df
```

### Step 7. legal_drinker 라는 Column을 만들고 다음과 같은 내용을 추가 합니다.
- 18세 이상은 True, 18세 미만은 False

```python
def legal(age):
  if age >= 18:
    return True
  else:
    return False
```

```python
df["legal_drinker"] = df["age"].apply(legal) # df["legal_drinker"] colunms 새로 만들어 내용을 추가하겠다!
```

```python
df["legal_drinker"] # 생성딘 것을 확인
```

### Step 8. 숫자로된 dataset에 10을 곱하세요
- 아무의미 없음

```python
def mul10(num_arg):
    if type(num_arg) == int: # 입력받은 객체의 type이 int면
        return num_arg * 10 # 객체에 10을 곱해라
    else:
        return num_arg # 그게 아니라면 그대로 num_arg를 출력해라
```

```python
df.applymap(mul10)
```
